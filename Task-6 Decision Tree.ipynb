# Importing required libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
import sklearn.datasets as datasets                #importing the  dataset from sklearn             
```

```python

iris=datasets.load_iris()            
x=pd.DataFrame(iris.data, columns=iris.feature_names)   #Creating dataframe with Pandas
pd.set_option('display.max_rows',150)                   #display maximum rows 
x                                                       #storing iris.data in x and executing it.
```
